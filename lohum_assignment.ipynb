```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import json

data = []
for line in open('/content/CRS-16.json', 'r'):
    data.append(json.loads(line))
```

```python
data=pd.json_normalize(data)
```

```python
data
```

```python
data['velocity'].unique()
```

```python
data['altitude'].unique().sum()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(data['altitude'])
plt.title("Distribution Of altitude",fontsize=20)
```

#Obseravtion
1-altitude is very high in the range of 200-210

2-altitude has the second high range between 0-10

3-between 10-200 altitude is having almost same value

```python
plt.figure(figsize=(12,8))
sns.distplot(data['velocity'])
plt.title("Distribution Of Velocity",fontsize=20)
```

#Obseravtion
1-Velocity is having a high value between 0-400

2-Velocity is also high between the range of 1500-2000

3-after 500 velocity is decreasing till 1500

4-after almost 2000 velocity is continuously decreasing

#Relationship between velocity and altitude

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=data['velocity'],y=data['altitude'])
plt.title("Relationship Between Velocity and Altitude",fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Altitude",fontsize=15)
plt.xlabel("Velocity",fontsize=15)
```

#Obseravtion
1-As the altitude is increasing the velocity is also increasing 

2-Same as the velocity will increase the altitude is going to increase

3-As when the altitude was 65 the velocity increased by 0-1800 approx but after reaching the 65 value on altitude the altitude is increasing very fast 

4-And after 4000 velocity no matter if we are increasing the velocity it is not crossing the altitude value of 200

#Rate of change

```python
data['rate of change']=data['altitude']/data['time']*100
```

```python
data
```

#Rate of change vs velocity

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
sns.distplot(data['rate of change'])
plt.subplot(1,2,2)
sns.distplot(data['velocity'])
plt.tight_layout()
```

#Obseravtion
1-there is a reverse relationship between upper two graphs like in velocity it is first increasing and then decreasing

2-and in rate of change it is increasing from a low value
